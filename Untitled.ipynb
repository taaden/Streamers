```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list subscribers based on countries
x_axis = ["US", "Brazil", "UK", "France", "Germany", "Mexico", "Canada", "Australia", "Argentina", "Japan"]

# Set the y-axis to subscribers. 
y_axis = [65.3, 17.9, 16.7, 8.6, 8.2, 8.1, 7.9, 6.9, 5.4, 4.2]
```

```python
plt.plot(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
```

```python
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
plt.plot(x_axis, y_axis)
plt.show()
```
